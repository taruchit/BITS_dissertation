```python
import matplotlib.pyplot as plt
```

```python
label_count=cic_df['Label'].value_counts()
f = plt.figure()
f.set_figwidth(50)
f.set_figheight(10)
plt.yscale("log")
plt.bar(label_count.index, label_count.values)
plt.xlabel('Type of event')
plt.ylabel('Frequency')
plt.title('Frequency of event type')
plt.show()
```

```python
classlabel_count=cic_df['ClassLabel'].value_counts()
f = plt.figure()
f.set_figwidth(50)
f.set_figheight(10)
plt.bar(classlabel_count.index, classlabel_count.values)
plt.xlabel('Type of event')
plt.ylabel('Frequency')
plt.title('Frequency of event type')
plt.show()
```
